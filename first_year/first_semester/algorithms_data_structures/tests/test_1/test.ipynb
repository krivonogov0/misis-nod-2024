Сделайте функцию, которая параметром будет принимать список и удалять из него все дубли.

````python
def remove_repeats(input_list: list) -> list:
    """
    Remove repeats in input list. 
    
    Parameters:
    -----------
    `input_list` : `list`
        The input list.

    Returns:
    --------
    `output_list` : `list`
        List without repeats.
    """
    output_list = list(set(input_list))

    return output_list

def main():
    list_with_repeats: list = [1, "a", 2, 2, "a", 3.0, "b", "3.0"]

    list_with_no_repeats = remove_repeats(list_with_repeats)

    print(f"Original list: {list_with_repeats}")
    print(f"Processed list: {list_with_no_repeats}")


if __name__ == "__main__":
    main()
````

#### **Задание 6.**

Сделайте функцию, которая будет возвращать сколько дней прошло или осталось до заданной даты в году, в зависимости от того, была уже эта дата или нет.

````python
from datetime import datetime
from typing import List

def get_difference_of_date(input_date: List[int]) -> int:
    """
    Get the difference in days of the specified date. 

    Parameters:
    -----------
    `input_date` : `List[int]`
        The input list with date (year, month, day).

    Returns:
    --------
    `difference` : `int`
        Difference in days.
    """
    date: datetime = datetime.today().replace(hour=0, minute=0, second=0, microsecond=0)
    target_date: datetime = datetime(input_date[0], input_date[1], input_date[2])

    diff_stamps = date.timestamp() - target_date.timestamp()

    difference: int = int(diff_stamps / (24 * 60 * 60))

    return difference

def main():
    target_date: List[int] = [2024, 10, 10]
    today_date: datetime = datetime.today().strftime("%d-%m-%Y")

    difference = get_difference_of_date(target_date)

    print(f"Today: {today_date}. Target: {target_date}. Difference: {difference} (minus sign means that the day has not yet arrived)")


if __name__ == "__main__":
    main()
````

Сделайте функцию, которая будет возвращать сколько дней осталось до конца текущего месяца.

````python
from datetime import datetime

def get_days_end_of_month() -> int:
    """
    Get the number of days until the end of the month. 

    Returns:
    --------
    `number_of_days` : `int`
        Number of days.
    """
    curr_day: int = datetime.today().day
    curr_month: int = datetime.today().month

    if curr_month == 2:
        last_day = 28
    elif curr_month in (4, 6, 9, 11):
        last_day = 30
    else:
        last_day = 31

    number_of_days: int = last_day - curr_day

    return number_of_days

def main():
    today_date: datetime = datetime.today().strftime("%d-%m-%Y")

    number_of_days = get_days_end_of_month()

    print(f"Today: {today_date}. Days until the end of the month: {number_of_days}")


if __name__ == "__main__":
    main()
````

````python
import random
from typing import Tuple, List

def random_fill_list(number: int, interval: Tuple[float]) -> List[float]:
    """
    Fill the list with N random numbers from a given interval. 

    Parameters:
    -----------
    `number` : `int`
        The number of random elements.
    `number` : `int`
        The given interval for random.

    Returns:
    --------
    `random_list` : `List[float]`
        Random list.
    """
    random_list: List[float] = [random.uniform(interval[0], interval[1]) for _ in range(number)]

    return random_list

def main():
    interval: Tuple[float] = (0.0, 123.321)
    number: int = 10

    random_list = random_fill_list(number, interval)

    print(f"Random list: {random_list}")


if __name__ == "__main__":
    main()
````

````python
from typing import List, Dict
from pprint import pprint

def refactor_list(input_list: List[Dict[str, str]]) -> Dict[int, Dict[int, List[str]]]:
    """
    Refactor list structure. 

    Parameters:
    -----------
    `input_list` : `List[Dict[str, str]]`
        The input list for refactoring.

    Returns:
    --------
    `event_dict` : `Dict[int, Dict[int, List[str]]]`
        Dict from refactored list.
    """
    event_dict: dict = dict()

    for event in input_list:
        year = int(event['date'].split('-')[0])
        month = int(event['date'].split('-')[1])
        
        if year not in event_dict:
            event_dict[year] = dict()
        
        if month not in event_dict[year]:
            event_dict[year][month] = list()

        event_dict[year][month].append(event['event'])

    return event_dict

def main():
    events = [
        {
            'date': '2019-12',
            'event': 'name1',
        },
        {
            'date': '2019-12',
            'event': 'name2',
        },
        {
            'date': '2019-11',
            'event': 'name3',
        },
        {
            'date': '2019-11',
            'event': 'name4',
        },
        {
            'date': '2020-10',
            'event': 'name5',
        },
        {
            'date': '2020-10',
            'event': 'name6',
        },
        {
            'date': '2020-11',
            'event': 'name5',
        },
        {
            'date': '2020-11',
            'event': 'name6',
        },
        {
            'date': '2020-12',
            'event': 'name7',
        },
        {
            'date': '2020-12',
            'event': 'name8',
        },
        {
            'date': '2020-12',
            'event': 'name9',
        },
    ]

    output_dict = refactor_list(events)

    pprint(output_dict)


if __name__ == "__main__":
    main()
````

```
[
```

````python
from typing import List

def get_directories(url: str) -> List[str]:
    """
    Get list of directories.

    Parameters:
    -----------
    `url` : `str`
        The input URL.

    Returns:
    --------
    `directories` : `List[str]`
        List with directories.
    """
    directories: list = list()

    folders = url.split('/')
    del folders[0], folders[0], folders[0], folders[-1]
    
    for i in range(len(folders)):
        path = str("/" + "/".join(folders[i:]) + "/")
        directories.append(path)

    return directories

def main():
    url: str = 'http://test.com/dir1/dir2/dir3/page.html'

    directories = get_directories(url)

    print(directories)


if __name__ == "__main__":
    main()
````

#### **Задание 14.**

Даны два числа. Выведите в консоль процесс умножения этих чисел в столбик, как в школе.

````python
def column_multiplication(num_one: int, num_two: int) -> None:
    """
    Print column multiplication.

    Parameters:
    -----------
    `num_one` : `int`
        The first number.
    `num_two` : `int`
        The second number.
    """
    if num_one < num_two:
        num_one, num_two = num_two, num_one

    str_num_one: str = str(num_one)
    str_num_two: str = str(num_two)

    max_length = max(len(str_num_one), len(str_num_two))
    diff = len(str_num_one) - len(str_num_two)

    print(' ' * (len(str_num_one) - diff), str_num_one)
    print(' ' * (len(str_num_two) + diff), str_num_two)
    print(' ' * (len(str_num_two)), '-' * max_length)

    multiplications: list = list()
    for idx in range(len(str_num_two)):
        multiplication = str(int(str_num_one) * int(str_num_two[-idx-1]))
        multiplications.append(multiplication)

    for i in range(len(multiplications)):
        if len(multiplications[i]) > max_length:
            print(' ' * (len(multiplications) - i - 1), multiplications[i])
        else:
            print(' ' * (len(multiplications) - i), multiplications[i])
    
    total_product = num_one * num_two
    print(' ', '-' * len(str(total_product)))
    print(' ', total_product)

def main():
    column_multiplication(6789, 123456)


if __name__ == "__main__":
    main()
````

#### **Задание 15.**

Даны два числа. Выведите в консоль процесс деления этих чисел в столбик, как в школе.

````python
def column_division(dividend: int, divisor: int) -> None:
    """
    Print column division.

    Parameters:
    -----------
    `dividend` : `int`
        The dividend.
    `divisor` : `int`
        The divisor.
    
    Raises:
    -------
    `ValueError`
        If the divisor equals zero.
    """
    if divisor == 0:
        raise ValueError(f"Division by zero!")

    str_dividend: str = str(dividend)

    result: str = ""
    current: str = ""
    index: int = 0
    len_dividend: int = len(str_dividend)

    padding: int = 4

    print(" " * padding + str_dividend)
    
    while index < len_dividend:
        current += str_dividend[index]
        index += 1

        if int(current) < divisor:
            if result:
                result += "0"
            continue

        quotient = int(current) // divisor
        result += str(quotient)

        product = quotient * divisor
        remainder = int(current) - product

        print(" " * (padding + len(result) - 1), product)
        print(" " * (padding + len(result) - 1), "-" * len(str(product)))

        if remainder != 0 or index < len_dividend:
            print(" " * (padding + len(result) - 1), remainder)

        current = str(remainder)

    print(" " * padding + "-" * max(len(str_dividend), len(result)))
    print(" " * padding + result)

def main():
    column_division(1234, 11)


if __name__ == "__main__":
    main()
````

#### **Задание 16.**

Дано число. Выведите на экран процесс разложения числа на простые множители, как в школе.

````python
def prime_factorization(number: int) -> None:
    """
    Print prime factorization.

    Parameters:
    -----------
    `number` : `int`
        The number.
    """
    factors: list = list()

    while number % 2 == 0:
        factors.append(2)
        number //= 2
    
    for i in range(3, int(number**0.5) + 1, 2):
        while number % i == 0:
            factors.append(i)
            number //= i

    if number > 2:
        factors.append(number)
    
    print(factors)


def main():
    prime_factorization(123456)


if __name__ == "__main__":
    main()
````

Напишите программу, которая будет находить корни квадратного уравнения.

Для этого спросите у пользователя три числа, которые будут коэффициентами уравнения.

````python
def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number

def solve_quadratic_equation(a: float, b: float, c: float) -> tuple:
    """
    Return solution of the quadratic equation.

    Parameters:
    -----------
    `a` : `int`
        The coefficient a.
    `b` : `int`
        The coefficient b.
    `c` : `int`
        The coefficient c.

    Returns:
    --------
    `solution` : `tuple`
        Solution of the quadratic equation.
    """
    discriminant = b ** 2 - 4 * a * c
    if discriminant < 0:
        solution: tuple = tuple()
        return solution
    
    x_first = round((-b - discriminant ** 0.5) / (2 * a), 3)
    x_second = round((-b + discriminant ** 0.5) / (2 * a), 3)

    solution: tuple = (x_first, x_second)
    
    return solution

def main():
    a = input("Input coefficient a: ")
    a = process_input(a)

    b = input("Input coefficient b: ")
    b = process_input(b)
    
    c = input("Input coefficient c: ")
    c = process_input(c)

    solution: tuple = solve_quadratic_equation(a, b, c)

    print(f"Корни уравнения {a}x^2 + {b}x + {c}: {solution}")
    
if __name__ == "__main__":
    main()
````

#### **Задание 18.**

Напишите код, который будет генерировать пароль заданного размера.

Пароль обязательно должен будет содержать маленькую и большую буквы, цифру и какой-то специальный символ.

````python
import string
import random

def generate_password(length: int) -> str:
    """
    Return password with specified length.

    Parameters:
    -----------
    `length` : `int`
        The length of the password.

    Returns:
    --------
    `password` : `str`
        Generated password.
    """
    lowercases: str = string.ascii_lowercase
    uppercases: str = string.ascii_uppercase
    digits: str = string.digits
    special_chars: str = string.punctuation

    symbols = lowercases + uppercases + digits + special_chars

    password_list = [symbols[random.randint(0, 93)] for _ in range(length - 4)]
    password_list += [random.choice(lowercases), random.choice(uppercases), random.choice(digits), random.choice(special_chars)]

    password = ''.join(password_list)
    
    return password

def main():
    length: int = 7
    
    password = generate_password(length)

    print(f"Пароль: {password}")
    
if __name__ == "__main__":
    main()
````

#### **Задание 19.**

Напишите код, который будет проверять сложность пароля. Критерии сложности придумайте самостоятельно.

````python
import string

def check_password(password: str) -> None:
    """
    Сheck the password complexity.

    Parameters:
    -----------
    `password` : `str`
        The password.
    """
    lowercases = string.ascii_lowercase
    uppercases = string.ascii_uppercase
    digits = string.digits
    special_chars = string.punctuation
    
    minimal_length: int = 10

    if len(password) < minimal_length:
        print(f"- Длина пароля менее {minimal_length} символов!")

    is_lowercases = any(char in lowercases for char in password)
    if not is_lowercases:
        print("- Пароль не содержит строчные буквы!")

    is_uppercases = any(char in uppercases for char in password)
    if not is_uppercases:
        print("- Пароль не содержит заглавные буквы!")
    
    is_digits = any(char in digits for char in password)
    if not is_digits:
        print("- Пароль не содержит цифры")
    
    is_special_chars = any(char in special_chars for char in password)
    if not is_special_chars:
        print("- Пароль не содержит специальные символы!")


def main():
    password = "7AAFFF"

    check_password(password)

    
if __name__ == "__main__":
    main()
````

#### **Задание 20.**

Реализуйте алгоритм [Решето Эратосфена](https://ru.wikipedia.org/wiki/Решето_Эратосфена) для поиска простых чисел в заданном промежутке.

````python
def sieve_of_eratosthenes(start: int, stop: int) -> list:
    """
    The "sieve of Eratosthenes" algorithm.

    Parameters:
    -----------
    `start` : `int`
        The beginning of the interval.
    `end` : `int`
        The end of the interval.

    Returns:
    --------
    `simple_digits` : `list`
        Simple digits.
    """
    if stop < 2:
        simple_digits: list = []
        return simple_digits

    simple_digits: list = [True] * (stop + 1)
    simple_digits[0], simple_digits[1] = False, False

    for s in range(2, int(stop ** 0.5) + 1):
        if simple_digits[s]:
            for i in range(s * s, stop + 1, s):
                simple_digits[i] = False

    simple_digits: list = [i for i in range(start, stop + 1) if simple_digits[i] and i >= start]

    return simple_digits

def main():
    start = 10
    stop = 50

    simple_digits = sieve_of_eratosthenes(start, stop)

    print(f"Простые числа в промежутке [{start};{stop}]: {simple_digits}")

    
if __name__ == "__main__":
    main()
````

#### **Задача 1**. Сумма первых n положительных чисел

Напишите программу, запрашивающую у пользователя число и подсчитывающую сумму натуральных положительных чисел от 1 до введенного пользователем значения.

Сумма первых n положительных чисел может быть рассчитана по формуле: 

*sum = [n * (n + 1)] / 2*

````python
def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number 

def calculate_sum(number: float) -> int:
    """
    Calculate the sum of natural positive numbers from 1 to number.

    Parameters:
    -----------
    `number` : `int`
        The number for calculation.

    Returns:
    --------
    `sum` : `int`
        The calculated sum.

    Raises:
    -------
    `ValueError`
        If the number is not greater then zero.
    """
    if number < 0:
        raise ValueError(f"Your number: {number} is less then zero! Please enter a number is greater than zero.") 
    
    number = int(number)

    sum = number * (number + 1) / 2
    sum = int(sum)

    return sum

def main():
    print("Input the number to get the sum of natural positive numbers from 1 to your number.")
    text_input = input("Number: ")
    print(f"Input is: {text_input}")

    number = process_input(text_input)
    sum = calculate_sum(number)

    print(f"The sum is: {sum}")


if __name__ == "__main__":
    main()
````

#### **Задача 3**.

Вычислите значение следующего выражения (аргументы - целые числа и вводятся с клавиатуры):

*f = [(((x ** 5) + 7) / (|-6| * y)) ** (1/3)] / [7 - z mod y]*

Округлите результат до 3-х знаков после запятой, используя функцию round().


````python
def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number 

def calculate_function_value(x: float, y: float, z: float) -> float:
    """
    Calculate the value of function f(x,y,z).

    Parameters:
    -----------
    `x` : `float`
        The first function argument.
    `y` : `float`
        The second function argument.
    `z` : `float`
        The third function argument.

    Returns:
    --------
    `result` : `float`
        The calculated value.

    Raises:
    -------
    `ValueError`
        - If the second argument equals zero (division by zero).
        - If the divider (7 - z mod y) is zero (division by zero).
        - If the number under the root is not greater than zero (only positive real numbers can be raised to a fractional rational power).
    """
    if y == 0:
        raise ValueError("Division by zero! Please enter another y value.")

    divider = 7 - z % y
    if divider == 0:
        raise ValueError("The divider (7 - z mod y) is zero! Please enter another pair of values (z, y).")

    number_under_the_root = ((x ** 5 + 7) / (6 * y))
    if number_under_the_root < 0:
        raise ValueError("The number under the root is not greater than zero! Please enter another pair of values (x, y).")
    
    value = (number_under_the_root ** (1/3)) / divider

    value = round(value, 3)

    return value

def main():
    print("Input three numbers to get the function value.")
    text_input_1 = input("Number 1: ")
    text_input_2 = input("Number 2: ")
    text_input_3 = input("Number 3: ")
    print(f"Input is: {text_input_1}, {text_input_2}, {text_input_3}")

    x = process_input(text_input_1)
    y = process_input(text_input_2)
    z = process_input(text_input_3)

    value = calculate_function_value(x, y, z)

    print(f"The function value is: {value}")


if __name__ == "__main__":
    main()
````

#### **Задача 4**.

Дана электрическая цепь, состоящая из 2-х последовательно соединенных проводников (сопротивление каждого известно).

Найти общее сопротивление цепи (округление результата необходимо выполнить до 1-го знака после запятой).

````python
def calculate_resistance(first_resistance: float, second_resistance: float) -> float:
    """
    Calculate the resistance of two conductors connected in series.

    Parameters:
    -----------
    `first_resistance` : `float`
        Resistance of the first element.
    `second_resistance` : `float`
        Resistance of the second element.

    Returns:
    --------
    `resistance` : `float`
        The calculated resistance.
    
    Raises:
    -------
    `ValueError`
        If any resistance is less than zero (not valid).
    """
    if first_resistance < 0 or second_resistance < 0:
        raise ValueError("The resistance is less than zero! Please enter another value of resistance.")

    resistance = first_resistance + second_resistance

    resistance = round(resistance, 1)

    return resistance

def main():
    print("Specify the resistance of the conductors (connected in series).")
    first_resistance = 10.3
    second_resistance = 23.241
    
    resistance = calculate_resistance(first_resistance, second_resistance)

    print(f"The resistance is: {resistance} (for resistances: {first_resistance}, {second_resistance})")


if __name__ == "__main__":
    main()
````

#### **Задача 5**.

Дано уравнение *ax + b = 0* и отрезок [m;n].

Ответьте на вопрос, попадает ли решение уравнения в указанный отрезок.

````python
def solve_equation(a: float, b: float) -> float:
    """
    Find a solution to the equation ax + b = 0.

    Parameters:
    -----------
    `a` : `float`
        The first parameter.
    `b` : `float`
        The second parameter.

    Returns:
    --------
    `x` : `float`
        The solution of the equation.
    
    Raises:
    -------
    `ValueError`
        If first parameter (a) equals zero (division by zero). 
    """
    if a == 0:
        raise ValueError("Division by zero! Please enter another value of 'a'.")

    x = -b / a
    x = round(x, 7)

    return x

def is_point_in_segment(x: float, m: float, n: float) -> bool:
    """
    Checks whether the point belongs to the segment [m;n].

    Parameters:
    -----------
    `x` : `float`
        The point.
    `m` : `float`
        The beginning of the segment.
    `n` : `float`
        The end of the segment.

    Returns:
    --------
    `is_in_segment` : `float`
        The boolean value (True if the point belongs to the segment).
    
    Raises:
    -------
    `ValueError`
        If end of the segment less then beginning of the segment. 
    """
    if n < m:
        raise ValueError("End of the segment less then beginning of the segment! Please enter another values.")

    if x >= m and x <= n:
        is_in_segment = True
    else:
        is_in_segment = False

    return is_in_segment

def main():
    print("Specify the values of the parmeters (ax + b = 0) and values of the segment [m;n].")
    a = -18
    b = -0.54

    m = -3
    n = 5
    
    x = solve_equation(a, b)
    answer = is_point_in_segment(x, m, n)

    print(f"Does the solution ({x}) belongs to the segment [{m},{n}]? {answer}")


if __name__ == "__main__":
    main()
````

#### **Задача 6**. Поездка по кругу

Спортсмен решил потренироваться перед марафоном и покататься вокруг города на скорость. Длина дороги — 123 километра. 

Спортсмен стартует с нулевого километра и едет со скоростью v километров в час. На какой отметке он остановится через t часов?

Реализуйте программу, которая спрашивает у пользователя v и t и выводит целое число от 0 до 122 — номер километра, на котором остановится Спортсмен. Учтите, что он может прокатиться больше одного круга.

````python
def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number 

def calculate_num_kilometer(v: float, t: float) -> int:
    """
    Сalculates the kilometer number where the sportsman will stop (the length of the lap is 123 kilometers).

    Parameters:
    -----------
    `v` : `float`
        The velocity of sportsman.
    `t` : `float`
        The duration of the ride.
    
    Returns:
    --------
    `kilometer` : `int`
        The kilometer number.

    Raises:
    -------
    `ValueError`
        - If the velocity or the time is less than zero.
    """
    if v < 0 or t < 0:
        raise ValueError("The velocity or the time is less than zero! Please enter valid values.")
        
    path_length = v * t
    kilometer = path_length % 123

    kilometer = int(kilometer)

    return kilometer

def main():
    print("Input velocity and time.")
    text_input_1 = input("Velocity: ")
    text_input_2 = input("Time: ")
    print(f"Velocity is: {text_input_1}, time is: {text_input_2}")

    v = process_input(text_input_1)
    t = process_input(text_input_2)

    kilometer = calculate_num_kilometer(v, t)

    print(f"The number of kilometer is: {kilometer}")


if __name__ == "__main__":
    main()
````

#### **Задача 7**. Сумма и произведение цифр в числе

Дано двузначное и трехзначное число. 

Для каждого выведите на экран сумму и произведение цифр.

````python
from typing import Tuple

def calculate_digits(number: int) -> Tuple[int, int]:
    """
    Сalculate the sum and product of digits.

    Parameters:
    -----------
    `number` : `int`
        Number for calculating.

    Returns:
    --------
    `result` : `Tuple[int, int]`
        The tuple with calculated sum and product of digits.
    """
    if number < 0:
        number = -number

    sum_of_digits = 0
    product_of_digits = 1

    while number > 0:
        digit = number % 10
        sum_of_digits += digit
        product_of_digits *= digit
        number //= 10

    return (sum_of_digits, product_of_digits)

def main():
    print("Specify the two-digit number and the three-digit number.")
    two_digit_number = 14
    three_digit_number = 293
    
    first_result = calculate_digits(two_digit_number)
    second_result = calculate_digits(three_digit_number)

    print(f"For two-digit number {two_digit_number} the result is (sum, product): {first_result}")
    print(f"For three-digit number {three_digit_number} the result is (sum, product): {second_result}")


if __name__ == "__main__":
    main()
````

#### **Задача 8**. Сортировка трех чисел

Напишите программу, запрашивающую у пользователя три целых числа и выводящую их в упорядоченном виде – по возрастанию. 

Используйте функции min и max для нахождения наименьшего и наибольшего значений. 

Оставшееся число можно найти путем вычитания из суммы трех введенных чисел максимального и минимального.

````python
from typing import Tuple

def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number 

def sort_three_numbers(num_1: float, num_2: float, num_3: float) -> Tuple[float, float, float]:
    """
    Sorts three numbers in ascending order.

    Parameters:
    -----------
    `num_1` : `float`
        The first number.
    `num_2` : `float`
        The second number.
    `num_1` : `float`
        The third number.
    
    Returns:
    --------
    `sorted_numbers` : `Tuple[float, float, float]`
        The tuple with sorted numbers.
    """
    numbers = (num_1, num_2, num_3)

    min_num = int(min(numbers))
    max_num = int(max(numbers))
    avg_num = int(sum(numbers) - min_num - max_num)

    return (min_num, avg_num, max_num)

def main():
    print("Input three numbers.")
    text_input_1 = input("Number 1: ")
    text_input_2 = input("Number 2: ")
    text_input_3 = input("Number 3: ")
    print(f"Your numbers are: {text_input_1}, {text_input_2}, {text_input_3}")

    num_1 = process_input(text_input_1)
    num_2 = process_input(text_input_2)
    num_3 = process_input(text_input_3)

    sorted_numbers = sort_three_numbers(num_1, num_2, num_3)

    print(f"Sorted numbers (min -> max): {sorted_numbers}")


if __name__ == "__main__":
    main()
````

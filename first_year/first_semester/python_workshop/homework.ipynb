Первый побочный закидывается в основной, там считается количество цифр 5, количество выводится на экран, и затем они удаляются из основного списка.

После этого в основной закидывается второй побочный список, там считается количество цифр 3 и выводится на экран. В конце также выводится и сам список.

Из интереса вы попросили вашего друга показать код его программы и поняли, что сделали это не зря — то, что вы увидели, повергло вас в шок и ужас. Вот сам код:

```python
a = [1, 5, 3]
b = [1, 5, 1, 5]
c = [1, 3, 1, 5, 3, 3]
for i in b:
    a.append(i)
t = 0
for i in a:
    if i == 5:
        t += 1
print(t)
d = []
for i in a:
    if i != 5:
        d.append(i)
for i in c:
    d.append(i)
```

#### **Задача 2.**

Два класса стоят в две отдельные шеренги. В каждом классе ученики выстроены по росту (по возрастанию): 

- в одном классе от 160 см до 176 см с шагом 2, 
- во втором классе — от 162 см до 180 см с шагом 3. 

Спустя какое-то время два класса объединяют в одну шеренгу и тоже выстраивают их по возрастанию.

Напишите программу, которая генерирует списки роста для каждого в классе, затем объединяет их в один список и сортирует его в порядке возрастания. Выведите отсортированный список на экран.

Отсортированный список учеников: [160, 162, …]

````python
def main():
````
````python
    guests = ['Петя', 'Ваня', 'Саша', 'Лиза', 'Катя']

    answer_options = ["Пришел", "пришел", "Ушел", "ушел", "Пора спать", "пора спать"]
    
    print("Party tracker.\n")

    while True:
        print(f"Сейчас на вечеринке {len(guests)} человек: {guests}")
        
        came_or_went = input("Гость пришел или ушел? (введите 'Пора спать' для окончания вечеринки): ")

        if came_or_went not in answer_options:
            print(f"Ответ '{came_or_went}' - некорректен! Попробуйте снова.")
        else:
            if came_or_went in ["Пора спать", "пора спать"]:
                print("\nВечеринка закончилась, все легли спать.")
                return 0
            
            guest_name = input("Имя гостя: ")

            if came_or_went in ["Пришел", "пришел"]:
                guest_name = guest_name.capitalize()
                if len(guests) < 6:
                    print(f"Привет, {guest_name}!")
                    guests.append(guest_name)
                else:
                    print(f"Прости, {guest_name}, но мест нет.")

            if came_or_went in ["Ушел", "ушел"]:
                guest_name = guest_name.capitalize()
                if guest_name in guests:
                    print(f"Пока, {guest_name}!")
                    guests.remove(guest_name)
                else:
                    print(f"Гостя '{guest_name}' - нет на вечеринке! Попробуйте снова.")


if __name__ == "__main__":
    main()
````

#### **Задача 5.**

Мы пишем приложение для удобного прослушивания музыки. 

У Вани есть список из девяти песен группы Depeche Mode. Каждая песня состоит из названия и продолжительности с точностью до долей минут:

```python
violator_songs = [
    ['World in My Eyes', 4.86],
    ['Sweetest Perfection', 4.43],
    ['Personal Jesus', 4.56],
    ['Halo', 4.9],
    ['Waiting for the Night', 6.07],
    ['Enjoy the Silence', 4.20],
    ['Policy of Truth', 4.76],
    ['Blue Dress', 4.29],
    ['Clean', 5.83]
]
```

Из этого списка Ваня хочет выбрать N песен и закинуть их в особый плейлист с другими треками. И при этом ему важно, сколько времени в сумме эти N песен будут звучать.

Напишите программу, которая запрашивает у пользователя количество песен из списка и затем названия этих песен, а на экран выводит общее время их звучания.

Пример:
```
Сколько песен выбрать? 3
Название 1-й песни: Halo
Название 2-й песни: Enjoy the Silence
Название 3-й песни: Clean


Общее время звучания песен: 14,93 минуты
```

````python
def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number 

def main():
    violator_songs = [
        ['World in My Eyes', 4.86],
        ['Sweetest Perfection', 4.43],
        ['Personal Jesus', 4.56],
        ['Halo', 4.9],
        ['Waiting for the Night', 6.07],
        ['Enjoy the Silence', 4.20],
        ['Policy of Truth', 4.76],
        ['Blue Dress', 4.29],
        ['Clean', 5.83]
    ]

    music_names = [music_batch[0] for music_batch in violator_songs]
    
    print("Music duration calculator.\n")

    music_number = input("Сколько песен выбрать? ")
    music_number = process_input(music_number)
    music_number = int(music_number)
    
    duration_time = 0

    for music_id in range(0, music_number):
        music_name = input(f"Название {music_id + 1}-й песни: ")
        if music_name in music_names:
            playlist_id = music_names.index(music_name)
            duration_time += violator_songs[playlist_id][1]
        else: 
            print(f"Нет такой песни - '{music_name}'! Доступные треки: {music_names}")

    print(f"Общее время звучания песен: {round(duration_time, 2)}")


if __name__ == "__main__":
    main()
````

#### **Задача 6.**

Даны два списка целых чисел, оба списка заполняются с клавиатуры.
В первый список вводится три числа, во второй — семь чисел. 

Напишите программу, которая запрашивает у пользователя эти числа, затем расширяет первый список элементами второго и после этого оставляет в первом списке только уникальные элементы, то есть удаляет лишние повторы чисел. 

Условный оператор использовать нельзя.

Пример:
```
Введите 1-е число для первого списка: 1
Введите 2-е число для первого списка: 2
Введите 3-е число для первого списка: 3
Введите 1-е число для второго списка: 2
Введите 2-е число для второго списка: 4
Введите 3-е число для второго списка: 6

Введите 6-е число для второго списка: 2
Введите 7-е число для второго списка: 1


Первый список: [1, 2, 3]
Второй список: [2, 4, 6, 3, 3, 2, 1]


Новый первый список с уникальными элементами: [4, 6, 3, 2, 1]
```

````python
def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number 

def main():
    first_list = list()
    second_list = list()

    print("Заполните 3 числа первого списка.")
    for i in range(0, 3):
        number = input(f"Введите {i+1}-е число для первого списка: ")
        number = process_input(number)
        number = int(number)
        first_list.append(number)
    
    print("Заполните 7 чисел второго списка.")
    for i in range(0, 7):
        number = input(f"Введите {i+1}-е число для второго списка: ")
        number = process_input(number)
        number = int(number)
        second_list.append(number)

    print(f"Первый список: {first_list}")
    print(f"Второй список: {second_list}")

    first_list.extend(second_list)
    first_list = list(set(second_list))

    print(f"\nНовый список с уникальными элементами: {first_list}")


if __name__ == "__main__":
    main()
````

#### **Задача 7.**

Частная контора даёт в прокат ролики самых разных размеров.

Человек может надеть ролики любого размера, которые не меньше размера его ноги. 

Пользователь вводит два списка размеров: N размеров коньков и K размеров ног людей.

Реализуйте код, который определяет, какое наибольшее число человек сможет одновременно взять ролики и пойти покататься.

Пример:
```
Кол-во коньков: 4
Размер 1-й пары: 41
Размер 2-й пары: 40
Размер 3-й пары: 39
Размер 4-й пары: 42


Кол-во людей: 3
Размер ноги 1-го человека: 42
Размер ноги 2-го человека: 41
Размер ноги 3-го человека: 42


Наибольшее кол-во людей, которые могут взять ролики: 2
```

````python
def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number 

def main():
    skates_list = list()
    foot_list = list()

    skates_number = input(f"Введите количество коньков: ")
    skates_number = process_input(skates_number)
    skates_number = int(skates_number)

    print("Заполните список размеров коньков.")
    for i in range(0, skates_number):
        number = input(f"Введите размер {i+1}-й пары: ")
        number = process_input(number)
        number = int(number)
        skates_list.append(number)

    foot_number = input(f"Введите количество людей: ")
    foot_number = process_input(foot_number)
    foot_number = int(foot_number)
    
    print("Заполните список размеров ног людей.")
    for i in range(0, foot_number):
        number = input(f"Введите размер ноги {i+1}-го человека: ")
        number = process_input(number)
        number = int(number)
        foot_list.append(number)

    skates_list.sort()
    foot_list.sort()

    max_num_customers = 0
    for foot_size in foot_list:
        for skate_size in skates_list:
            if foot_size <= skate_size:
                skates_list.remove(skate_size)
                max_num_customers += 1

    print(f"\nНаибольшее кол-во людей, которые могут взять ролики: {max_num_customers}")


if __name__ == "__main__":
    main()
````

#### **Задача 8.**

N человек, пронумерованных числами от 1 до N, стоят в кругу. 

Они начинают играть в считалку на выбывание, где каждый K-й по счёту человек выбывает из круга, после чего счёт продолжается со следующего за ним человека.

На вход подаётся количество человек N и номер K. Напишите программу, которая выводит число от 1 до N — это номер человека, который останется в кругу последним.

Пример:
```
Кол-во человек: 5
Какое число в считалке? 7
Значит, выбывает каждый 7-й человек


Текущий круг людей: [1, 2, 3, 4, 5]
Начало счёта с номера 1
Выбывает человек под номером 2


Текущий круг людей: [1, 3, 4, 5]
Начало счёта с номера 3

Текущий круг людей: [1, 3, 4]
Начало счёта с номера 1
Выбывает человек под номером 1


Текущий круг людей: [3, 4]
Начало счёта с номера 3
Выбывает человек под номером 3


Остался человек под номером 4
```

````python
def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number 

def main():
    num_people = input(f"Введите кол-во человек: ")
    num_people = process_input(num_people)
    num_people = int(num_people)

    people_circle = [people_id + 1 for people_id in range(0, num_people)]

    num_repeat = input(f"Введите число в считалке: ")
    num_repeat = process_input(num_repeat)
    num_repeat = int(num_repeat)

    out_id = 0

    while len(people_circle) > 1:
        print(f"\nТекущий круг людей: {people_circle}")
        
        out_id = out_id % (len(people_circle))
        print(f"Начало счета с номера: {people_circle[out_id]}")
        
        out_id = (num_repeat + out_id - 1) % (len(people_circle))
        print(f"Выбывает человек под номером: {people_circle[out_id]}")

        del people_circle[out_id]

    print(f"\n\nОстался человек под номером: {people_circle[0]}")


if __name__ == "__main__":
    main()
````

````python
def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number 

def main():
    num_friends = input(f"Введите кол-во друзей: ")
    num_friends = process_input(num_friends)
    num_friends = int(num_friends)

    balance = [0 for _ in range(num_friends)]

    num_receipt = input(f"Введите кол-во долговых расписок: ")
    num_receipt = process_input(num_receipt)
    num_receipt = int(num_receipt)

    for receipt_id in range(num_receipt):
        print(f"Введите данные {receipt_id + 1}-й расписки.")

        to_whom = input(f"Кому: ")
        to_whom = process_input(to_whom)
        to_whom = int(to_whom)
        
        from_whom = input(f"От кого: ")
        from_whom = process_input(from_whom)
        from_whom = int(from_whom)
    
        how_much = input(f"Сколько: ")
        how_much = process_input(how_much)
        how_much = int(how_much)

        balance[to_whom - 1] -= how_much
        balance[from_whom - 1] += how_much

    print("\nБаланс друзей:")
    for friend_num, friend_balance in enumerate(balance):
        print(f"{friend_num}: {friend_balance}")


if __name__ == "__main__":
    main()
````

#### **Задача 10.**

Последовательность чисел называется симметричной, если она одинаково читается как слева направо, так и справа налево.

Например, следующие последовательности являются симметричными:

Пользователь вводит последовательность из N чисел.

Напишите программу, которая определяет, какое минимальное количество и каких чисел надо приписать в конец этой последовательности, чтобы она стала симметричной.

```
Последовательность: [1, 2, 1, 2, 2]
```

Пример 2:

````python
def process_input(text_input: str) -> float:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number 

def main():
    num_numbers = input(f"Введите кол-во чисел: ")
    num_numbers = process_input(num_numbers)
    num_numbers = int(num_numbers)

    numbers = list()

    for number in range(num_numbers):
        number = input(f"Число: ")
        number = process_input(number)
        number = int(number)

        numbers.append(number)

    print(f"Последовательность: {numbers}")

    additional_numbers = list()

    n = len(numbers)
    is_found_prefix = False
    for i in range(n):
        prefix = numbers[:i]
        suffix = numbers[i:]
        if suffix == suffix[::-1]:
            additional_numbers = prefix[::-1]
            is_found_prefix = True
            break
    
    if not is_found_prefix:
        additional_numbers = numbers[:-1][::-1]

    print(f"Нужно приписать чисел: {len(additional_numbers)}")
    print(f"Добавочные числа: {additional_numbers}")


if __name__ == "__main__":
    main()
````

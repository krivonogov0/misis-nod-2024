#### **Упражнение 1.**

Дан список из чисел.

Определите их [НОК](https://ru.wikipedia.org/wiki/Наименьшее_общее_кратное) (наименьшее общее кратное) и [НОД](https://ru.wikipedia.org/wiki/Наибольший_общий_делитель) (наибольший общий делитель).

````python
import random
from typing import List

def nod_two_numbers(a: int, b: int) -> int:
    """
    Compute NOD of two numbers. 
    
    Auxiliary function for compute NOD of list.

    Parameters:
    -----------
    `a` : `int`
        The first number.
    `b` : `int`
        The second number.

    Returns:
    --------
    `nod_value` : `int`
        NOD of two numbers.
    """
    # Euclid's algorithm
    while a > 0 and b > 0:
        if a >= b:
            a = a % b
        else:
            b = b % a

    nod_value = max(a,b)

    return nod_value

def nok_two_numbers(a: int, b: int) -> int:
    """
    Compute NOK of two numbers. 
    
    Auxiliary function for compute NOK of list.

    Parameters:
    -----------
    `a` : `int`
        The first number.
    `b` : `int`
        The second number.

    Returns:
    --------
    `nok_value` : `int`
        NOK of two numbers.
    """
    nok_value = abs(a * b) // nod_two_numbers(a, b)
    
    return nok_value

def nod_list(numbers: List[int]) -> int:
    """
    Compute NOD of list of numbers. 
    
    Parameters:
    -----------
    `numbers` : `list[int]`
        The list of numbers

    Returns:
    --------
    `nod_of_list` : `int`
        NOD of list.
    """
    nod_of_list = numbers[0]
    
    for num in numbers[1:]:
        nod_of_list = nod_two_numbers(nod_of_list, num)

    return nod_of_list

def nok_list(numbers: List[int]) -> int:
    """
    Compute NOK of list of numbers. 
    
    Parameters:
    -----------
    `numbers` : `list[int]`
        The list of numbers

    Returns:
    --------
    `nok_of_list` : `int`
        NOK of list.
    """
    nok_of_list = numbers[0]
    
    for num in numbers[1:]:
        nok_of_list = nok_two_numbers(nok_of_list, num)

    return nok_of_list

def main():
    numbers: List[int] = [random.randint(1, 10) for _ in range(random.randint(2, 10))]
    
    nok = nok_list(numbers)
    nod = nod_list(numbers)

    print(f"Список чисел: {numbers}")
    print(f"НОК: {nok}")
    print(f"НОД: {nod}")


if __name__ == "__main__":
    main()
````

#### **Упражнение 2.**

Дан список из чисел.

Даны n предложений. Определите, сколько из них содержат хотя бы одну цифру.

````python
import random
from typing import List

def is_sentence_with_num(sentence: str) -> bool:
    """
    Сhecks whether the sentence contains a number.

    Parameters:
    -----------
    `sentence` : `str`
        The sentence.

    Returns:
    --------
    `is_with_num` : `bool`
        If it contains, then it is True.
    """
    is_with_num: bool = False

    words: list = list()
    words = sentence.split()
    for word in words:
        try:
            _ = int(word)
            is_with_num = True
        except:
            pass

    return is_with_num

def main():
    n = 10

    subjects = ["Кот", "Пес", "Человек", "Программист", "Актер"]
    verbs = ["бежит", "прыгает", "танцует", "разрабатывает", "пишет"]
    objects = ["в парке", "на улице", "на сцене", "код", "историю"]
    adverbs = ["быстро", "медленно", "грустно", "весело", "интенсивно"]

    sentences: List[str] = list()

    for _ in range(n):
        random_subject = random.choice(subjects)
        random_verb = random.choice(verbs)
        random_object = random.choice(objects)
        random_adverb = random.choice(adverbs)

        is_add_number = random.choice([True, False])

        if is_add_number:
            numbers: List[str] = []
            num_of_numbers = random.randint(0, 5)
            
            for _ in range(num_of_numbers):
                number = str(random.randint(0, 100))
                numbers.append(number)

            numbers_str = " ".join(numbers)
            sentence = f"{random_subject} {random_verb} {numbers_str} {random_object} {random_adverb}."

        else:
            sentence = f"{random_subject} {random_verb} {random_object} {random_adverb}."

        sentences.append(sentence)
    
    num_sentences = 0
    for sentence in sentences:
        is_contains_num = is_sentence_with_num(sentence)
        if is_contains_num:
            num_sentences += 1

    print(f"Список предложений: {sentences}")
    print(f"Из них содержат числа следующее число предложений: {num_sentences}")


if __name__ == "__main__":
    main()
````

#### **Упражнение 4.**

Для введенного предложения выведите статистику символ=количество. 

Регистр букв не учитывается.

````python
from typing import Dict

def character_statistics(sentence: str) -> Dict[str, int]:
    """
    Calculates statistics on the count of each character in a sentence.

    Parameters:
    -----------
    `sentence` : `str`
        The sentence.

    Returns:
    --------
    `characters` : `Dict[str, int]`
        The dictionary containing characters and their count in a sentence.
    """    
    characters: Dict[str, int] = dict()

    sentence = sentence.lower()

    for char_id in range(len(sentence)):
        char = sentence[char_id]

        if char not in characters:
            characters[char] = 1
        else:
            characters[char] += 1

    return characters

def main():
    sentence = input("Введите предложение: ")
    print(f"Sentence: {sentence}")

    characters = character_statistics(sentence)

    print(f"Statistics: ")
    for char, count in characters.items():
        print(f"{char} = {count}")
    

if __name__ == "__main__":
    main()
````

#### **Упражнение 5.**

Используя шифр [Цезаря](https://ru.wikipedia.org/wiki/Шифр_Цезаря) (достаточно только букв русского алфавита, знаки препинания не изменяются), зашифруйте, а затем расшифруйте введенную строку.

````python
def caesar_encrypter(string: str, shift: int) -> str:
    """
    Encrypts the string with the Caesar cipher.

    Parameters:
    -----------
    `string` : `str`
        The string.
    `shift` : `str`
        The cipher shift.

    Returns:
    --------
    `encrypted_string` : `str`
        The encrypted string.
    """
    russian_alphabet = [chr(i) for i in range(ord('а'), ord('я') + 1)]
    shifted_russian_alphabet = russian_alphabet[shift::] + russian_alphabet[:shift:]

    encrypted_string: str = ""

    string = string.lower()
    for char_id in range(len(string)):
        char = string[char_id]

        if char in russian_alphabet:
            origin_char_idx = russian_alphabet.index(char)
            shifted_char = shifted_russian_alphabet[origin_char_idx]
            encrypted_string += shifted_char
        else:
            encrypted_string += char

    return encrypted_string

def caesar_decrypter(string: str, shift: int) -> str:
    """
    Decrypts the string with the Caesar cipher.

    Parameters:
    -----------
    `string` : `str`
        The string.
    `shift` : `str`
        The cipher shift.

    Returns:
    --------
    `decrypted_string` : `str`
        The decrypted string.
    """    
    russian_alphabet = [chr(i) for i in range(ord('а'), ord('я') + 1)]
    shifted_russian_alphabet = russian_alphabet[shift::] + russian_alphabet[:shift:]

    decrypted_string: str = ""

    string = string.lower()
    for char_id in range(len(string)):
        char = string[char_id]

        if char in russian_alphabet:
            shifted_char_idx = shifted_russian_alphabet.index(char)
            unshifted_char = russian_alphabet[shifted_char_idx]
            decrypted_string += unshifted_char
        else:
            decrypted_string += char

    return decrypted_string

def main():
    string = input("Введите строку: ")
    print(f"Введенная строка: {string}")

    encrypted_string = caesar_encrypter(string=string, shift=3)
    print(f"Зашифрованная строка: {encrypted_string}")
    decrypted_string = caesar_decrypter(string=encrypted_string, shift=3)
    print(f"Расшифрованная строка: {decrypted_string}")


if __name__ == "__main__":
    main()
````

#### **Упражнение 6.**

Напишите функцию, которая принимает неограниченное количество числовых аргументов и возвращает кортеж из двух списков:

* отрицательных значений (отсортирован по убыванию);
* неотрицательных значений (отсортирован по возрастанию).

````python
import random
from typing import Tuple

def sort_negative_positive(numbers: list) -> Tuple[list, list]:
    """
    Sorts the numbers into two lists:
        * negative values (sorted in descending order);
        * non-negative values (sorted in ascending order).

    Parameters:
    -----------
    `numbers` : `list`
        The list of numbers.

    Returns:
    --------
    `sorted_numbers` : `Tuple[list, list]`
        Sorted values.
    """
    negative_numbers: list = list()
    positive_numbers: list = list()

    for number in numbers:
        if number < 0:
            negative_numbers.append(number)
        else:
            positive_numbers.append(number)

    negative_numbers.sort(reverse=True)
    positive_numbers.sort()
    
    sorted_numbers = (negative_numbers, positive_numbers)

    return sorted_numbers

def main():
    numbers: list = [random.randint(-100, 100) for _ in range(random.randint(1, 20))]

    sorted_numbers = sort_negative_positive(numbers)

    print(f"Числовые аргументы: {numbers}")
    print(f"Отсортированные значения: {sorted_numbers}")


if __name__ == "__main__":
    main()
````

#### **Упражнение 7.**

Строка называется палиндромом, если она пишется одинаково в обоих направлениях.

Например, палиндромами в английском языке являются слова «anna», «civic», «level», «hannah». 

Напишите программу, запрашивающую у пользователя строку и при помощи цикла определяющую, является ли она палиндромом.

````python
def is_palindrome(string: str) -> bool:
    """
    Checks whether the string is a palindrome.

    Parameters:
    -----------
    `string` : `str`
        The string.

    Returns:
    --------
    `is_str_palindrome` : `bool`
        If the string is a palindrome then the value is True.
    """
    is_str_palindrome: bool = True

    string = string.lower()

    for char_id in range(len(string)):
        char = string[char_id]
        char_reverse = string[-char_id-1]
        if char != char_reverse:
            is_str_palindrome = False
            break

    return is_str_palindrome

def main():
    input_string = input("Введите строку: ")
    print(f"Введенная строка: {input_string}")

    is_str_palindrome = is_palindrome(input_string)
    print(f"Является ли строка палиндромом: {is_str_palindrome}")

if __name__ == "__main__":
    main()
````

#### **Упражнение 8.**

Вы загадываете число от 1 до 100 (включительно). 

Компьютер спрашивает у вас «Твое число равно, меньше или больше, чем число N?»,  где N — число, которое хочет проверить компьютер.

Вы отвечаете одним из трёх чисел:

* 1 — равно,
* 2 — больше,
* 3 — меньше.

Напишите программу, которая с помощью цепочки таких вопросов и ответов угадывает число.

Дополнительно: сделайте так, чтобы можно было гарантированно угадать число за семь попыток.

````python
def process_input(text_input: str) -> int:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number or not from 1 to 3).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
        number = int(number)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")
    
    if number not in [1, 2, 3]:
        raise ValueError(f"Your input: '{text_input}' does not match the allowed! Please enter a number from 1 to 3.")

    return number

def main():
    print("Guess the number from 0 to 100 in no more than 7 attempts.\n")

    low: int = 0
    high: int = 100

    for attempt_num in range(7):
        attempt_num += 1
        hidden_number = (low + high) // 2

        print(f"Is your number is {hidden_number}?")

        text_input = input("Выбор (1, 2, 3): ")
        choice = process_input(text_input)

        print(f"- Your choice is: {choice}")

        if choice == 2:
            low = hidden_number + 1

        if choice == 3:
            high = hidden_number - 1

        if choice == 1:
            print(f"\nThe hidden number is guessed in {attempt_num-1} attempts!")
            print(f"The hidden number is {hidden_number}!")
            break


if __name__ == "__main__":
    main()
````

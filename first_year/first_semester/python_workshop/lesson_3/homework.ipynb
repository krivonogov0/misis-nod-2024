#### **Упражнение 1.**

Дан список из чисел.

Определите их [НОК](https://ru.wikipedia.org/wiki/Наименьшее_общее_кратное) (наименьшее общее кратное) и [НОД](https://ru.wikipedia.org/wiki/Наибольший_общий_делитель) (наибольший общий делитель).

````python
import random
from typing import List

def nod_two_numbers(a: int, b: int) -> int:
    """
    Compute NOD of two numbers. 
    
    Auxiliary function for compute NOD of list.

    Parameters:
    -----------
    `a` : `int`
        The first number.
    `b` : `int`
        The second number.

    Returns:
    --------
    `nod_value` : `int`
        NOD of two numbers.
    """
    # Euclid's algorithm
    while a > 0 and b > 0:
        if a >= b:
            a = a % b
        else:
            b = b % a

    nod_value = max(a,b)

    return nod_value

def nok_two_numbers(a: int, b: int) -> int:
    """
    Compute NOK of two numbers. 
    
    Auxiliary function for compute NOK of list.

    Parameters:
    -----------
    `a` : `int`
        The first number.
    `b` : `int`
        The second number.

    Returns:
    --------
    `nok_value` : `int`
        NOK of two numbers.
    """
    nok_value = abs(a * b) // nod_two_numbers(a, b)
    
    return nok_value

def nod_list(numbers: List[int]) -> int:
    """
    Compute NOD of list of numbers. 
    
    Parameters:
    -----------
    `numbers` : `list[int]`
        The list of numbers

    Returns:
    --------
    `nod_of_list` : `int`
        NOD of list.
    """
    nod_of_list = numbers[0]
    
    for num in numbers[1:]:
        nod_of_list = nod_two_numbers(nod_of_list, num)

    return nod_of_list

def nok_list(numbers: List[int]) -> int:
    """
    Compute NOK of list of numbers. 
    
    Parameters:
    -----------
    `numbers` : `list[int]`
        The list of numbers.

    Returns:
    --------
    `nok_of_list` : `int`
        NOK of list.
    """
    nok_of_list = numbers[0]
    
    for num in numbers[1:]:
        nok_of_list = nok_two_numbers(nok_of_list, num)

    return nok_of_list

def main():
    numbers: List[int] = [random.randint(1, 10) for _ in range(random.randint(2, 10))]
    
    nok = nok_list(numbers)
    nod = nod_list(numbers)

    print(f"Список чисел: {numbers}")
    print(f"НОК: {nok}")
    print(f"НОД: {nod}")


if __name__ == "__main__":
    main()
````

#### **Упражнение 4.**

Для введенного предложения выведите статистику символ=количество. 

Регистр букв не учитывается.

````python
from typing import Dict

def character_statistics(sentence: str) -> Dict[str, int]:
    """
    Calculates statistics on the count of each character in a sentence.

    Parameters:
    -----------
    `sentence` : `str`
        The sentence.

    Returns:
    --------
    `characters` : `Dict[str, int]`
        The dictionary containing characters and their count in a sentence.
    """    
    characters: Dict[str, int] = dict()

    sentence = sentence.lower()

    for char_id in range(len(sentence)):
        char = sentence[char_id]

        if char not in characters:
            characters[char] = 1
        else:
            characters[char] += 1

    return characters

def main():
    sentence = input("Введите предложение: ")
    print(f"Sentence: {sentence}")

    characters = character_statistics(sentence)

    print(f"Statistics: ")
    for char, count in characters.items():
        print(f"{char} = {count}")
    

if __name__ == "__main__":
    main()
````

#### **Упражнение 5.**

Используя шифр [Цезаря](https://ru.wikipedia.org/wiki/Шифр_Цезаря) (достаточно только букв русского алфавита, знаки препинания не изменяются), зашифруйте, а затем расшифруйте введенную строку.

````python
def caesar_encrypter(string: str, shift: int) -> str:
    """
    Encrypts the string with the Caesar cipher.

    Parameters:
    -----------
    `string` : `str`
        The string.
    `shift` : `str`
        The cipher shift.

    Returns:
    --------
    `encrypted_string` : `str`
        The encrypted string.
    """
    russian_alphabet = [chr(i) for i in range(ord('а'), ord('я') + 1)]
    shifted_russian_alphabet = russian_alphabet[shift::] + russian_alphabet[:shift:]

    encrypted_string: str = ""

    string = string.lower()
    for char_id in range(len(string)):
        char = string[char_id]

        if char in russian_alphabet:
            origin_char_idx = russian_alphabet.index(char)
            shifted_char = shifted_russian_alphabet[origin_char_idx]
            encrypted_string += shifted_char
        else:
            encrypted_string += char

    return encrypted_string

def caesar_decrypter(string: str, shift: int) -> str:
    """
    Decrypts the string with the Caesar cipher.

    Parameters:
    -----------
    `string` : `str`
        The string.
    `shift` : `str`
        The cipher shift.

    Returns:
    --------
    `decrypted_string` : `str`
        The decrypted string.
    """    
    russian_alphabet = [chr(i) for i in range(ord('а'), ord('я') + 1)]
    shifted_russian_alphabet = russian_alphabet[shift::] + russian_alphabet[:shift:]

    decrypted_string: str = ""

    string = string.lower()
    for char_id in range(len(string)):
        char = string[char_id]

        if char in russian_alphabet:
            shifted_char_idx = shifted_russian_alphabet.index(char)
            unshifted_char = russian_alphabet[shifted_char_idx]
            decrypted_string += unshifted_char
        else:
            decrypted_string += char

    return decrypted_string

def main():
    string = input("Введите строку: ")
    print(f"Введенная строка: {string}")

    encrypted_string = caesar_encrypter(string=string, shift=3)
    print(f"Зашифрованная строка: {encrypted_string}")
    decrypted_string = caesar_decrypter(string=encrypted_string, shift=3)
    print(f"Расшифрованная строка: {decrypted_string}")


if __name__ == "__main__":
    main()
````

#### **Упражнение 6.**

Напишите функцию, которая принимает неограниченное количество числовых аргументов и возвращает кортеж из двух списков:

* отрицательных значений (отсортирован по убыванию);
* неотрицательных значений (отсортирован по возрастанию).

````python
import random
from typing import Tuple

def sort_negative_positive(numbers: list) -> Tuple[list, list]:
    """
    Sorts the numbers into two lists:
        * negative values (sorted in descending order);
        * non-negative values (sorted in ascending order).

    Parameters:
    -----------
    `numbers` : `list`
        The list of numbers.

    Returns:
    --------
    `sorted_numbers` : `Tuple[list, list]`
        Sorted values.
    """
    negative_numbers: list = list()
    positive_numbers: list = list()

    for number in numbers:
        if number < 0:
            negative_numbers.append(number)
        else:
            positive_numbers.append(number)

    negative_numbers.sort(reverse=True)
    positive_numbers.sort()
    
    sorted_numbers = (negative_numbers, positive_numbers)

    return sorted_numbers

def main():
    numbers: list = [random.randint(-100, 100) for _ in range(random.randint(1, 20))]

    sorted_numbers = sort_negative_positive(numbers)

    print(f"Числовые аргументы: {numbers}")
    print(f"Отсортированные значения: {sorted_numbers}")


if __name__ == "__main__":
    main()
````

````python
def is_palindrome(string: str) -> bool:
    """
    Checks whether the string is a palindrome.

    Parameters:
    -----------
    `string` : `str`
        The string.

    Returns:
    --------
    `is_str_palindrome` : `bool`
        If the string is a palindrome then the value is True.
    """
    is_str_palindrome: bool = True

    string = string.lower()

    for char_id in range(len(string)):
        char = string[char_id]
        char_reverse = string[-char_id-1]
        if char != char_reverse:
            is_str_palindrome = False
            break

    return is_str_palindrome

def main():
    input_string = input("Введите строку: ")
    print(f"Введенная строка: {input_string}")

    is_str_palindrome = is_palindrome(input_string)
    print(f"Является ли строка палиндромом: {is_str_palindrome}")

if __name__ == "__main__":
    main()
````

````python
def process_input(text_input: str) -> int:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number or not from 1 to 3).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
        number = int(number)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")
    
    if number not in [1, 2, 3]:
        raise ValueError(f"Your input: '{text_input}' does not match the allowed! Please enter a number from 1 to 3.")

    return number

def main():
    print("Guess the number from 0 to 100 in no more than 7 attempts.\n")

    low: int = 0
    high: int = 100

    for attempt_num in range(7):
        attempt_num += 1
        hidden_number = (low + high) // 2

        print(f"Is your number is {hidden_number}?")

        text_input = input("Выбор (1, 2, 3): ")
        choice = process_input(text_input)

        print(f"- Your choice is: {choice}")

        if choice == 2:
            low = hidden_number + 1

        if choice == 3:
            high = hidden_number - 1

        if choice == 1:
            print(f"\nThe hidden number is guessed in {attempt_num-1} attempts!")
            print(f"The hidden number is {hidden_number}!")
            break


if __name__ == "__main__":
    main()
````

#### **Упражнение 9.**

Напишите программу, которая позволит пользователю преобразовывать числа из одной системы счисления в другую произвольным образом. 

Ваша программа должна поддерживать все системы счисления в диапазоне от 2 до 16 как для входных, так и для выходных данных. 

Если пользователь выберет систему с основанием, выходящим за границы допустимого, на экран должна быть выведена ошибка. 

Разделите код программы на несколько функций, включая функцию, конвертирующую число из произвольной системы счисления в десятичную, и обратную функцию, переводящую значение из десятичной системы в произвольную. 

В основной программе необходимо запросить у пользователя исходную систему счисления, целевую систему, а также число для преобразования.

````python
def process_input(text_input: str) -> int:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number or not from 2 to 16).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
        number = int(number)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")
    
    if number not in range(2, 17):
        raise ValueError(f"Your input: '{text_input}' does not match the allowed! Please enter a number from 2 to 16.")

    return number

def convert_into_decimal(number: str, system_base: int) -> float:
    """
    Converts a number from an arbitrary number system to a decimal one.

    Parameters:
    -----------
    `number` : `str`
        The number to convert.
    `system_base` : `int`
        The basis of the original number system.

    Returns:
    --------
    `decimal_number` : `float`
        The converted number to a decimal number system.
    
    Raises:
    -------
    `ValueError`
        If the user input is not valid (a number system cannot contain such a symbol).
    """
    digits = "0123456789ABCDEF"
    allowed_symbols = digits[:system_base] + ".-"
    symbols = set(number.upper())

    for symbol in symbols:
        if symbol not in allowed_symbols:
            raise ValueError(f"This number system cannot contain such a symbol: {symbol}! Allowed symbols: {allowed_symbols}.")

    if "-" in symbols:
        number = number.lstrip("-")
        is_negative = True
    else:
        is_negative = False

    integer_part, _, fractional_part = number.partition(".")

    decimal_number: float = 0.0

    for idx, char in enumerate(integer_part[::-1]):
        decimal_number += digits.index(char) * (system_base ** idx)

    for idx, char in enumerate(fractional_part):
        decimal_number += digits.index(char) * (system_base ** -(idx + 1))

    if is_negative:
        decimal_number *= -1

    return round(decimal_number, 10)

def convert_into_arbitrary(number: float, system_base: int) -> str:
    """
    Converts a number from a decimal number system to an arbitrary one.

    Parameters:
    -----------
    `number` : `float`
        The number to convert.
    `system_base` : `int`
        The basis of the target number system.

    Returns:
    --------
    `arbitrary_number` : `str`
        The converted number to an arbitrary number system.
    """
    digits = "0123456789ABCDEF"
    
    is_negative = number < 0
    if is_negative:
        number = -number

    integer_part: int = int(number)
    fractional_part: float = round((number - integer_part), 10)

    int_reminders = []
    while integer_part:
        int_reminders.append(digits[integer_part % system_base])
        integer_part //= system_base

    if not int_reminders:
        int_reminders.append('0')
        
    int_reminders.reverse()
    int_reminders = ''.join(int_reminders)

    float_wholes = []
    iteration = 0
    while fractional_part > 0 and iteration < 10:
        fractional_part *= system_base
        whole_part = int(fractional_part)
        float_wholes.append(digits[whole_part])
        fractional_part -= whole_part
        iteration += 1

    float_wholes = ''.join(float_wholes)
    
    arbitrary_number = int_reminders
    if float_wholes:
        arbitrary_number += '.' + float_wholes
    
    if is_negative:
        arbitrary_number = '-' + arbitrary_number

    return arbitrary_number

def main():
    print("The number system converter.\n")

    text_input_1 = input("Введите исходную систему счисления: ")
    origin_num_system = process_input(text_input_1)
    text_input_2 = input("Введите целевую систему счисления: ")
    target_num_system = process_input(text_input_2)
    text_input_3 = input("Введите число для преобразования: ")
    number = str(text_input_3)

    print(f"Исходная система счисления: {origin_num_system}")
    print(f"Целевая система счисления: {target_num_system}")
    print(f"Введенное число: {number}")

    decimal_number = convert_into_decimal(number, system_base=origin_num_system)
    arbitary_number = convert_into_arbitrary(decimal_number, system_base=target_num_system)

    print(f"\nЧисло в десятичной системе: {decimal_number}")
    print(f"\nПреобразованное число: {arbitary_number}")


if __name__ == "__main__":
    main()
````

#### **Упражнение 10.**

Магическими называются даты, в которых произведение дня и месяца составляет последние две цифры года.

Например, 10 июня 1960 года – магическая дата, поскольку 10 ´ 6 = 60.

Напишите функцию, определяющую, является ли введенная дата магической.

Используйте написанную функцию в главной программе для отображения всех магических дат в XX ве­ке.

````python
from typing import List
from pprint import pprint

def process_input(text_input: str) -> int:
    """
    Check the input text and convert it into a number.

    Parameters:
    -----------
    `text_input` : `str`
        The input text from user.

    Returns:
    --------
    `number` : `float`
        The number from user input.

    Raises:
    -------
    `ValueError`
        If the user input is not valid (empty or not a number).
    """
    if not len(text_input) > 0:
        raise ValueError(f"Your input: '{text_input}' is empty! Please enter a number.")
    
    try:
        number = float(text_input)
        number = int(number)
    except ValueError:
        raise ValueError(f"Your input: '{text_input}' is not a number! Please enter a number.")

    return number

def is_leap_year(year: int) -> bool:
    """Check is year is leap"""
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

def generate_dates_20th_century() -> List[List[int]]:
    """
    Generates dates in 20th century.

    Returns:
    --------
    `dates_list` : `List[List[int]]`
        List of dates.
    """ 
    months_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    dates_list = []

    for year in range(1900, 2000):
        for month in range(1, 13):
            days_in_month = months_days[month - 1]

            if month == 2 and is_leap_year(year):
                days_in_month = 29
            
            for day in range(1, days_in_month + 1):
                dates_list.append([day, month, year])
                
    return dates_list

def is_date_magical(date_list: List[int]) -> bool:
    """
    Checks if the date is 'magic'.

    Parameters:
    -----------
    `date_list` : `List[int]`
        Three components of the date.

    Returns:
    --------
    `is_magical` : `bool`
        If the date is magical, then the value is True.
    
    Raises:
    -------
    `ValueError`
        If the date is invalid (the value of the date, month, or year is incorrect).
    """ 
    date: int = date_list[0]
    month: int = date_list[1]
    year: int = date_list[2]

    if date < 1 or date > 31:
        raise ValueError(f"Date: {date} - is invalid! Please enter a different date.")
    if month < 1 or month > 12:
        raise ValueError(f"Month: {month} - is invalid! Please enter a different month.")
    if year < 0:
        raise ValueError(f"Year: {year} - is invalid! Please enter a different year.")

    is_magical: bool = False
    
    product_date_month: int = date * month
    if product_date_month == int(str(year)[-2:]):
        is_magical = True

    return is_magical


def main():
    print("Magical Date.\n")

    text_input_1 = input("Введите число: ")
    date = process_input(text_input_1)
    text_input_2 = input("Введите месяц (числом - не буквами!): ")
    month = process_input(text_input_2)
    text_input_3 = input("Введите год: ")
    year = process_input(text_input_3)

    date_list: List[int] = [date, month, year]
    is_magical = is_date_magical(date_list)

    print(f"Введенная дата: {date_list[0]}-{date_list[1]}-{date_list[2]}")
    print(f"Введенная магическая: {is_magical}")

    print("\nMagical Dates in XX:")
    xx_dates = generate_dates_20th_century()

    magical_dates_xx: list = list()
    for date_list in xx_dates:
        is_magical = is_date_magical(date_list)
        if is_magical: magical_dates_xx.append(date_list)
    
    pprint(magical_dates_xx)


if __name__ == "__main__":
    main()
````
